```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


data1=pd.read_csv('corna.csv')
data1.head(8)
```

```python
data1.info()
```

```python
Deaths_count_country=pd.DataFrame(data1.groupby(['Country'])['Deaths'].sum())
Deaths_count_country['Confirmed_counts']=pd.DataFrame(data1.groupby(['Country'])['Confirmed'].sum())


top10=Deaths_count_country.sort_values('Confirmed_counts',ascending=False).head(15)
x=pd.merge(top10,data1,on='Country')

Most_Deaths_country = x.drop_duplicates(subset=['Country'])

new_data1=Most_Deaths_country[['Country','Confirmed_counts','Deaths_x']]
new_data1
```

```python
print(data1.isnull().sum().to_frame('nulls'))
```

```python
len(data1)
```

```python
data=data1[data1.State.notnull()]
len(data)
```

```python
Deaths_count=pd.DataFrame(data.groupby(['State'])['Deaths'].sum())
Deaths_count['Recovered_counts']=pd.DataFrame(data.groupby(['State'])['Recovered'].sum())
Deaths_count['Confirmed_counts']=pd.DataFrame(data.groupby(['State'])['Confirmed'].sum())


top10=Deaths_count.sort_values('Confirmed_counts',ascending=False).head(30)
x=pd.merge(top10,data,on='State')

Most_Deaths = x.drop_duplicates(subset=['State'])

new_data=Most_Deaths[['State','Confirmed_counts','Deaths_x','Recovered_counts','Country']]
new_data
```

```python
new_data=new_data.iloc[0:10,:]

x=new_data['Confirmed_counts']
y=new_data['State']
plt.rcParams['figure.figsize'] = (15, 10)

sns.barplot(x,y,order=y )  #graf çizdir (Most popular)
```

```python
#çember
labels=new_data['State']
sizes=new_data['Confirmed_counts']
explode = None  # explode 1st slice
plt.subplots(figsize=(8,8))
plt.pie(sizes, explode=explode, labels=labels,autopct='%1.1f%%', shadow=False, startangle=0)
plt.axis('equal')
plt.show()
```

```python

Confirmed_count_date=pd.DataFrame(data1.groupby(['Last Update'])['Confirmed'].sum())


labels=Confirmed_count_date.index
sizes=Confirmed_count_date['Confirmed']
explode = None  # explode 1st slice
plt.subplots(figsize=(8,8))
plt.pie(sizes, explode=explode, labels=labels,autopct='%1.1f%%', shadow=False, startangle=0)
plt.axis('equal')
plt.show()
Confirmed_count_date
```
